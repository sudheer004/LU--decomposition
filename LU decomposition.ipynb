```python
import numpy as np

def lu_decomposition(matrix):
    rows, cols = matrix.shape
    # Create matrices L and U filled with zeros
    L = np.zeros((rows, cols))
    U = np.zeros((rows, cols))

    for j in range(cols):
        # U matrix
        U[:, j] = matrix[:, j]
        for i in range(j):
            # L matrix
            L[i, j] = matrix[i, j] / U[i, i]
            U[:, j] -= L[i, j] * U[:, i]

    # Fill diagonal of L with ones
    np.fill_diagonal(L, 1)

    return L, U

# Example usage with a 6x6 matrix
A = np.array([[2, -3, 4, 5, -7, 8],
              [-2, 0, -7, 6, 1, 9],
              [4, 1, -9, 7, 8, -2],
              [6, -7, 8, -9, 1, 3],
              [-4, 3, -2, 1, 0, -1],
              [-6, -5, 4, -3, -2, 7]])

L, U = lu_decomposition(A)

print("Matrix A:")
print(A)
print("\nMatrix L:")
print(L)
print("\nMatrix U:")
print(U)
```
